```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

```python
#school_data_complete.count()
school_data_complete.head()
```

```python
# Find total number of schools by unique name
school_total = len(school_data_complete['school_name'].unique())
#school_total

student_total = len(school_data_complete)
#student_total

budget_total = school_data_complete["budget"].unique().sum()
#budget_total

avg_math = school_data_complete['math_score'].mean()
#avg_math

avg_reading = school_data_complete['reading_score'].mean()
#avg_reading

# calulate passing math scores
math_pass = school_data_complete.loc[school_data_complete["math_score"] >= 70, :]
#print(math_pass)
math_pass_count = len(math_pass)
math_pass_percent = 100 * math_pass_count / student_total
#math_pass_percent

read_pass = school_data_complete.loc[school_data_complete["reading_score"] >= 70, :]
# print(read_pass)
read_pass_count = len(read_pass)
read_pass_percent = 100 * read_pass_count / student_total
#read_pass_percent


overall_pass = math_pass.loc[math_pass["reading_score"] >= 70, :]
#overall_pass
overall_pass_count = len(overall_pass)
overall_pass_percent = 100 * overall_pass_count / student_total
#overall_pass_percent

school_summary_df = pd.DataFrame({
    "Total Schools": [school_total],
    "Total Students": [student_total],
    "Total Budget": [budget_total],
    "Average Math Score": [avg_math],
    "Average Reading Score": [avg_reading],
    "% Passing Math": math_pass_percent,
    "% Passing Reading": read_pass_percent,
    "% Overall Passing": overall_pass_percent})

#format dataframe
school_summary_df["Total Budget"] = school_summary_df["Total Budget"].astype(float).map("${:,.2f}".format)
school_summary_df["Total Students"] = school_summary_df["Total Students"].astype(float).map("{:,}".format)

school_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# variables to hold results

school_type = school_data_complete.groupby('school_name')['type'].unique()
#school_type.head()
total_students = school_data_complete.groupby('school_name')['student_name'].count()
#total_students.head()
total_school_budget = school_data_complete.groupby('school_name')['budget'].unique()
#total_school_budget
per_student_budget = total_school_budget / total_students
#per_student_budget
avg_math = school_data_complete.groupby('school_name')['math_score'].mean()
#avg_math
avg_reading = school_data_complete.groupby('school_name')['reading_score'].mean()
#avg_reading

#variables to calculate passing math with 70 or higher
pass_math = school_data_complete.loc[school_data_complete['math_score'] >= 70]
pass_math_count = pass_math.groupby('school_name')['math_score'].count()
pass_math_100 = pass_math_count * 100
pass_math_percent = pass_math_100 / total_students
#pass_math_percent

#variables to calculate passing reading with 70 or higher
pass_read = school_data_complete.loc[school_data_complete['reading_score'] >= 70]
pass_read_count = pass_read.groupby('school_name')['reading_score'].count() 
pass_read_100 = pass_read_count * 100
pass_read_percent = pass_read_100 / total_students 
#pass_read_percent

#variables to calculate overall passing by checking for reading while already passing math
overall_pass = pass_math.loc[pass_math['reading_score'] >= 70]
overall_pass_count = overall_pass.groupby('school_name')['reading_score'].count()
overall_pass_100 = overall_pass_count * 100
overall_pass_percent = overall_pass_100 / total_students
#overall_pass_percent
```

```python
# summary dataframe
school_summary = pd.DataFrame({
                                "School Type" : school_type, 
                                "Total Students" : total_students, 
                                "Total School Budget" : total_school_budget,
                                "Per Student Budget" : per_student_budget,
                                "Average Math Score" : avg_math,
                                "Average Reading Score" : avg_reading,
                                "% Passing Math": pass_math_percent,
                                "% Passing Reading" : pass_read_percent,
                                "% Overall Passing" : overall_pass_percent})

# format summary dataframe

school_summary["Total School Budget"] = school_summary['Total School Budget'].astype(float).map("${:,.2f}".format)
school_summary['Per Student Budget'] = school_summary['Per Student Budget'].astype(float).map("${:.2f}".format)

school_summary
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school

```python
# Obtain and calculate 9th grade math score averages by school_name
ninth_grade = student_data.loc[student_data["grade"] == "9th"].groupby("school_name", as_index=False)
ninth_math_avg = pd.DataFrame(ninth_grade["math_score"].mean())
#ninth_math_avg

# Obtain and calculate 10th grade math score averages by school_name
tenth_grade = student_data.loc[student_data["grade"] == "10th"].groupby("school_name", as_index=False)
tenth_math_avg = pd.DataFrame(tenth_grade["math_score"].mean())
#tenth_math_avg

# Obtain and calculate 11th grade math score averages by school_name
eleventh_grade = student_data.loc[student_data["grade"] == "11th"].groupby("school_name", as_index=False)
eleventh_math_avg = pd.DataFrame(eleventh_grade["math_score"].mean())
#eleventh_math_avg

# Obtain and calculate 12th grade math score averages by school_name
twelveth_grade = student_data.loc[student_data["grade"] == "12th"].groupby("school_name", as_index=False)
twelveth_math_avg = pd.DataFrame(twelveth_grade["math_score"].mean())
#twelveth_math_avg
```

```python
# merge above grade scores to summary table
# merge 9th and 10th
math_grades = pd.merge(ninth_math_avg, tenth_math_avg, on="school_name")
# add 11th
math_grades = pd.merge(math_grades, eleventh_math_avg, on="school_name")
# add 12th
math_grades = pd.merge(math_grades, twelveth_math_avg, on="school_name")

# clean column headers
math_grades.columns = ["", "9th", "10th", "11th", "12th"]
math_grades
```
